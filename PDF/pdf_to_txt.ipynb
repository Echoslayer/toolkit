```python
import os
from PyPDF2 import PdfReader

pdf_number = "06"

# Convert the PDF to .txt format
# pdf_path = f'/home/wizard/Documents/lab file/Back to Basics/{pdf_number}.pdf'
pdf_path = f'G:/我的雲端硬碟/Course/研一上/實驗室資訊/資料/Back to Basics/{pdf_number}.pdf'

def pdf_to_txt(pdf_path, txt_path):
    with open(pdf_path, 'rb') as file:
        # 使用 PdfReader 代替 PdfFileReader
        pdf = PdfReader(file)
        with open(txt_path, 'w', encoding="utf-8") as output_file:
            for page_num in range(len(pdf.pages)):
                text = pdf.pages[page_num].extract_text()
                output_file.write(text)
    print(f"File {txt_path} created successfully!")

def read_txt_file(output_folder, pdf_number):
    """讀取txt檔案並回傳內容"""
    
    # 將 total_text.txt 儲存在指定的資料夾中
    txt_path = os.path.join(output_folder, f'{pdf_number}_total_text.txt')
    pdf_to_txt(pdf_path, txt_path)

    with open(txt_path, 'r', encoding="utf-8") as file:
        content = file.read()
    # Check if the file was successfully created
    os.path.exists(txt_path)

    return content

def folder_create(pdf_path, pdf_number):
    # 獲取pdf檔名，並創建相同名稱的資料夾
    dir_name = os.path.dirname(pdf_path)
    output_folder = os.path.join(dir_name, pdf_number)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    return output_folder

def split_txt_by_headers(content, headers, output_folder):  
    
    start_indices = [content.find(header) for header in headers]
    end_indices = start_indices[1:] + [len(content)]
    
    split_contents = []
    for start, end in zip(start_indices, end_indices):
        # 如果起始位置存在，則切割該段落
        if start != -1:
            split_contents.append(content[start:end])

    for header, split_content in zip(headers, split_contents):
        output_path = os.path.join(output_folder, f"{header}.txt")
        with open(output_path, 'w', encoding="utf-8") as file:
            file.write(split_content)

output_folder = folder_create(pdf_path, pdf_number)
print(output_folder)

total_content = read_txt_file(output_folder, pdf_number)

headers = ["REVIEW AND PURPOSE", "OBLIQUE INTERFERENCE OF TWO BEAMS", "ASSUMPTIONS"]

split_txt_by_headers(total_content, headers, output_folder)
```

```python
import os
pdf_number = "10"
file_path = f'G:/我的雲端硬碟/Course/實驗室資訊/資料/Back to Basics/chatgpt_md/{pdf_number}.md'

def folder_create(pdf_path, pdf_number):
    # 獲取pdf檔名，並創建相同名稱的資料夾
    dir_name = os.path.dirname(pdf_path)
    output_folder_path = os.path.join(dir_name, pdf_number)
    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)
    return output_folder_path

# Initialize empty lists to store the responses from humans and assistants
human_responses = []
assistant_responses = []

# Flags to indicate the current speaker
is_human = False
is_assistant = False

# Read the file and separate the responses based on the speaker
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        # Detect the speaker and set the flags accordingly
        if 'You:' in line:
            is_human = True
            is_assistant = False
            continue  # Skip the current line as it's a speaker identifier
        elif 'ChatGPT:' in line:
            is_human = False
            is_assistant = True
            continue  # Skip the current line as it's a speaker identifier

        # Append the responses to the appropriate list based on the current speaker
        if is_human:
            human_responses.append(line)
        elif is_assistant:
            assistant_responses.append(line)

# Concatenate the responses into single strings for each speaker
human_text = ''.join(human_responses)
assistant_text = ''.join(assistant_responses)

output_folder_path = folder_create(file_path, pdf_number)

# Save the separated responses to .txt files
human_file_path = f'{output_folder_path}/human_responses.md'
assistant_file_path = f'{output_folder_path}/assistant_responses.md'

with open(human_file_path, 'w', encoding='utf-8') as f:
    f.write(human_text)

with open(assistant_file_path, 'w', encoding='utf-8') as f:
    f.write(assistant_text)

human_file_path, assistant_file_path
```
