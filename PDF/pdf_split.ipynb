```python
from PyPDF2 import PdfReader, PdfWriter
import os

def extract_pages(input_pdf, page_numbers, output_pdf):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    for page_num in page_numbers:
        writer.add_page(reader.pages[page_num - 1])  # 頁數從0開始，所以需要減1

    with open(output_pdf, "wb") as out:
        writer.write(out)

# 使用範例
dir_path = r'F:\Code\DATA\ENGR 211 Solutions'
input_pdf = "M01_HIBB0000_00_SE_C01.pdf"  # 輸入的PDF文件名
pdf_path = os.path.join(dir_path, input_pdf)
page_numbers = [3, 8, 10, 15, 17]  # 要提取的頁數
output_pdf = f"hw_{input_pdf}"  # 輸出的PDF文件名
output_path = os.path.join(dir_path, output_pdf)

extract_pages(pdf_path, page_numbers, output_path)
```

```python
from PyPDF2 import PdfReader, PdfWriter
import os

def merge_pdfs(input_pdfs, output_pdf):
    writer = PdfWriter()

    for input_pdf in input_pdfs:
        reader = PdfReader(input_pdf)
        for page in reader.pages:
            writer.add_page(page)

    with open(output_pdf, "wb") as out:
        writer.write(out)

def extract_pages(input_pdf, page_numbers, output_pdf):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    for page_num in page_numbers:
        writer.add_page(reader.pages[page_num - 1])

    with open(output_pdf, "wb") as out:
        writer.write(out)

dir_path = r'F:\Code\DATA\ENGR 211 Solutions'

# 特殊处理：合并C04的三个部分
c04_parts = ['M01_HIBB0000_00_SE_C04_Part1.pdf', 'M01_HIBB0000_00_SE_C04_Part2.pdf', 'M01_HIBB0000_00_SE_C04_Part3.pdf']
c04_paths = [os.path.join(dir_path, part) for part in c04_parts]
merged_c04_path = os.path.join(dir_path, 'M01_HIBB0000_00_SE_C04.pdf')
merge_pdfs(c04_paths, merged_c04_path)

# 处理其他文件
for input_pdf, page_numbers in hw_dir.items():
    pdf_path = os.path.join(dir_path, input_pdf)
    output_pdf = f"hw_{input_pdf}"
    output_path = os.path.join(dir_path, output_pdf)
    extract_pages(pdf_path, page_numbers, output_path)
    print(f"Extracted pages from {input_pdf} to {output_pdf}")
```
