```python
import tensorflow as tf  # 如果您安裝了 TensorFlow-GPU
# 或者
# import torch  # 如果您安裝了 PyTorch

# 檢查是否有可用的 GPU
gpu_devices = tf.config.experimental.list_physical_devices('GPU')
if gpu_devices:
    print("有可用的 GPU。")
    for gpu in gpu_devices:
        print("GPU 名稱:", gpu.name)
else:
    print("沒有可用的 GPU。")
```
