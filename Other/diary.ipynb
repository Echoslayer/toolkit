```python
class Task:
    def __init__(self, date, todo=None):
        if todo is None:
            todo = {}
        self.todo = todo
        self.date = date.date()
        self.week = date.strftime("%A")

    def add_todo(self, todo):
        self.todo.update(todo)

def print_task(task: Task) -> None:
    title = f' - {task.date}({task.week})'
    todo_text = "\n    - ToDo"

    for todo, color in task.todo.items():
        todo_text += "\n"
        if color != "":
            todo_text += f"        - [ ] <span style='color:{color};'>{todo}</span> \n"
        else:
            todo_text += f"        - [ ] {todo}\n"
    todo_text += "    - Others\n"
    print(title + todo_text)


import datetime
date = datetime.datetime.now()
days = 7
tasks = []

for i in range(days):
    current_date = date + datetime.timedelta(days=i)
    if current_date.strftime("%A") == "Monday":
        todo = {
            "Yoga Class": "Silver",
            "Engineering Mechanics- Statics HW review": ""
        }
    elif current_date.strftime("%A") == "Tuesday":
        todo = {
            "Yoga Class": "Silver",
            "Engineering Mechanics- Statics HW review": ""
        }
    elif current_date.strftime("%A") == "Wednesday":
        todo = {
            "Yoga Class": "Silver",
            "Engineering Mechanics- Statics HW review": ""
        }
    elif current_date.strftime("%A") == "Thursday":
        todo = {
            "Yoga Class": "Silver",
            "Engineering Mechanics- Statics HW review": ""
        }
    elif current_date.strftime("%A") == "Friday":
        todo = {
            "Yoga Class": "Silver",
        }
    elif current_date.strftime("%A") == "Saturday":
        todo = {}
    else:
        todo = {}
    task = Task(current_date, todo)
    tasks.append(task)

for task in tasks:
    print_task(task)
```
