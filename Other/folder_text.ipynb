## 資料夾文件名稱

```python
import os
import re

def list_folder_contents(folder_path):
    """列出資料夾中的所有名稱

    Args:
        folder_path (_type_): 路徑

    Returns:
        _type_: _description_
    """
    
    name_list = []
    try:
        # 使用 os.listdir() 函數列出資料夾中的所有名稱
        contents = os.listdir(folder_path)
        # 列印所有名稱
        for item in contents:         
            if item != "desktop.ini":
                print(item)
                name_list.append(item)
            else:
                pass
    except OSError as e:
        print(f"發生錯誤: {e}")

    return name_list

def sort_names_by_alphabet(name_list):
    '''對檔案名稱進行排序並列印排序後的檔案名稱'''
    sorted_names = sorted(name_list)
    for name in sorted_names:
        print(name)

def extract_header_number(name):
    '''從檔案名稱中提取標頭數字部分，保留非數字部分為字符串'''
    match = re.search(r'(\d+(\.\d+)?)', name)
    if match:
        return match.group()
    else:
        return name

def sort_names_by_header(name_list):
    '''按照檔案名稱中的標頭數字部分進行排序並列印排序後的檔案名稱'''
    sorted_names = sorted(name_list, key=extract_header_number)
    for name in sorted_names:
        print(name)

def list_folder_contents(folder_path):
    name_list = []
    try:
        # 使用 os.listdir() 函數列出資料夾中的所有名稱
        contents = os.listdir(folder_path)
        # 列印所有名稱
        for item in contents:
            print(item)
            name_list.append(item)
    except OSError as e:
        print(f"發生錯誤: {e}")

    return name_list

def sort_names_by_number(name_list):
    '''按照檔案名稱中的數字部分進行排序並列印排序後的檔案名稱'''
    # 使用過濾器排除 desktop.ini 檔案名稱
    filtered_names = [name for name in name_list if not name.startswith('desktop.ini')]
    # 將檔案名稱按數字部分排序
    sorted_names = sorted(filtered_names, key=lambda x: float(x.split('_')[0]))
    # for name in sorted_names:
    #     print(name)
    return sorted_names
```

```python
def pdf_find_title(file_path):
    whole_text = []
    with open(file_path, 'rb') as pdf_file:
        # 創建 PDF 閱讀器對象
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        # 獲取 PDF 中的總頁數
        num_pages = len(pdf_reader.pages)

        # 遍歷每一頁並提取文本
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            page_text = page.extract_text()
            whole_text.append(page_text)
    
    # 假設您想查找以"Part"開頭的第一行文本
    titel = None
    for lines in whole_text:
        for line in lines.split('\n'):
            if line.startswith("Part"):
                titel = line
                break  # 找到後跳出循環
        if titel is not None:
            break  # 找到後跳出外部循環
    return titel
```

```python
import PyPDF2
import re

def pdf_read_text(filename):
    whole_text = ""
    with open(filename, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            page_text = page.extract_text()
            whole_text += page_text 
            whole_text += "\n"
    return whole_text.split('\n')

def reformat_sentences(filename):
    pdf_lines = pdf_read_text(filename)
    
    current_section = ""
    sections = {}
    section_id = None

    # 正則表達式來匹配可能的標題（連續的大寫字母和空格）
    title_pattern = re.compile(r'\b([A-Z][A-Z\s]*[A-Z])\b')

    for line in pdf_lines:
        # 尋找匹配的標題
        match = title_pattern.search(line)
        if match:
            # 如果已經有一個節標題，將其內容保存到sections字典中
            if section_id is not None:
                sections[section_id] = current_section
                current_section = ""
            section_id = match.group().strip()  # 設置新的節標題
        else:
            current_section += line + ' '

    # 添加最後一個節的內容
    if section_id is not None:
        sections[section_id] = current_section

    return sections

# for section_id, content in sections.items():
#     print(f"Section: {section_id}\nContent: {content}\n")
```

```python
filename = './Back to Basics/39.pdf'
sections = reformat_sentences(filename)


section_ids = list(sections.keys())
contents = list(sections.values())
for section_id, content in sections.items():
    print(f"Section: {section_id}\nContent: {content}\n")
```
